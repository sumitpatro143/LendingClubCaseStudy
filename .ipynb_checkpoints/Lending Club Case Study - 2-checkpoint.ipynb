Bucketing Different Variables For Better Analysis

```python
# Bucketing Loan Amount

def bucket_loan_amnt(column):
    
    if column <= 5000:
        return '0 - 5K'                         # 25% Quartile
    elif (column > 5000) and (column <= 10000):
        return '5K - 10K'
    elif (column > 10000) and (column <= 15000):
        return '10K - 15K'
    else:
        return '15K - above'                    # 75% Quartile

loan_df['loan_amnt_b'] = loan_df.apply(lambda x : bucket_loan_amnt(x['loan_amnt']), axis=1)
```

```python
# Bucketing Funded Amount

def bucket_funded_amnt(column):
    
    if column <= 5000:
        return '0 - 5K'                         # 25% Quartile
    elif (column > 5000) and (column <= 10000):
        return '5K - 10K'
    elif (column > 10000) and (column <= 15000):
        return '10K - 15K'
    else:
        return '15K - above'                    # 75% Quartile

loan_df['funded_amnt_b'] = loan_df.apply(lambda x : bucket_loan_amnt(x['funded_amnt']), axis=1)
```

```python
# Bucketing Funded Amount Investors

def bucket_funded_amnt_inv(column):
    
    if column <= 5000:
        return '0 - 5K'                         # 25% Quartile
    elif (column > 5000) and (column <= 10000):
        return '5K - 10K'
    elif (column > 10000) and (column <= 15000):
        return '10K - 15K'
    else:
        return '15K - above'                    # 75% Quartile

loan_df['funded_amnt_inv_b'] = loan_df.apply(lambda x : bucket_loan_amnt(x['funded_amnt_inv']), axis=1)
```

```python
# Bucketing Annual Income

def bucket_annual_inc(column):
    if column <= 40000:
        return '0 - 40K'                          # 25% Quartile
    elif (column > 40000) and (column <= 50000):
        return '40K - 50K'
    elif (column > 50000) and (column <= 60000):
        return '50K - 60K'
    elif (column > 60000) and (column <= 70000):
        return '60K - 70K'
    elif (column > 70000) and (column <= 80000):
        return '70K - 80K'
    else:
        return '80K - above'                      # 75% Quartile

loan_df['annual_inc_b'] = loan_df.apply (lambda x : bucket_annual_inc(x['annual_inc']), axis = 1)
```

```python
# Bucketing Interest Rate

def bucket_int_rate(column):
    if column <= 9:
        return 'Very Low'                  # 25% Quartile
    elif (column > 9) and (column <= 11):
        return 'Low'
    elif (column > 11) and (column <= 13):
        return 'Medium'
    elif (column > 13) and (column <= 15):
        return 'High'
    else:
        return 'Very High'                # 75% Quartile
    
loan_df['int_rate_b'] = loan_df.apply(lambda x : bucket_int_rate(x.int_rate), axis = 1)
```

```python
# Bucketing dti (debt to income ratio)

def bucket_dti (column):
    if column <= 8:
        return 'Very Low'                   # 25% Quartile
    elif (column > 8) and (column <= 12):
        return 'Low'
    elif (column < 12) and (column <= 16):
        return 'Medium'
    elif (column < 16) and (column <= 20):
        return 'High'
    else:
        return 'Very High'                  # 75% Quartile

loan_df['dti_b'] = loan_df.apply(lambda x : bucket_dti(x.dti), axis = 1)
```

From above boxplots we can see there are outliers for columns Loan Amount, Funded Amount, Funded Amount Investors, Interest Rate, Installment and Annual Income.

```python
# Lets remove the outliers

outliers_columns = ['Loan Amount', 'Funded Amount', 'Funded Amount Investors', 'Interest Rate', 'Installment', 'Annual Income']

def remove_outliers_iqr(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data [(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data

loan_df= remove_outliers_iqr(loan_df, outliers_columns)
```

### 7. ORDERED CATEGORICAL VARIABLE ANALYSIS

```python
# Plotting Pie Chart for Term

loan_term_counts = loan_df['Term'].value_counts()
colors = ['skyblue', 'lightgreen', 'lightcoral']

def fancy_barplot(x, y, data, **kwargs):
    ax = plt.gca()
    palette = sns.color_palette('pastel', len(data))
    
    for i, (value, color) in enumerate(zip(y, palette)):
        ax.bar(x=i, height=value, color=color, width=0.8, edgecolor='white', linewidth=2)
        ax.text(i, value + 0.05, f'{value}', ha='center', va='bottom', color='black', fontweight='bold')

# Pie Chart
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.pie(loan_term_counts, labels=loan_term_counts.index, autopct='%1.1f%%', colors=colors, shadow=True)
font1 = {'family':'cursive','color':'brown','size':20}
plt.title('Loan Term Distribution', fontdict=font1)


# Countplot
plt.subplot(1, 2, 2)
sns.set(style="whitegrid")
ax = sns.countplot(x='Term',data=loan_df,order=loan_df['Term'].value_counts().index)
font1 = {'family':'cursive','color':'brown','size':20}
font2 = {'family':'serif','color':'black','size':15}
plt.title("Count Plot : Term", fontdict=font1)
plt.xlabel("Name Of Term", fontdict=font2)
plt.ylabel("Number Of Applications Counts", fontdict=font2)

fancy_barplot(ax.get_xticks(), loan_df['Term'].value_counts(), loan_df['Term'], ax=ax)

plt.tight_layout()
plt.show()
```
